```python
# fonte: https://richardpricejones.medium.com/drawing-a-rectangle-with-a-angle-using-opencv-c9284eae3380
def draw_angled_rec(x0, y0, width, height, angle, cor, tam, img):
  import math
  _angle = angle * math.pi / 180.0
  b = math.cos(_angle) * 0.5
  a = math.sin(_angle) * 0.5
  pt0 = (int(x0 - a * height - b * width),
          int(y0 + b * height - a * width))
  pt1 = (int(x0 + a * height - b * width),
          int(y0 - b * height - a * width))
  pt2 = (int(2 * x0 - pt0[0]), int(2 * y0 - pt0[1]))
  pt3 = (int(2 * x0 - pt1[0]), int(2 * y0 - pt1[1]))

  cv2.line(img, pt0, pt1, cor, tam)
  cv2.line(img, pt1, pt2, cor, tam)
  cv2.line(img, pt2, pt3, cor, tam)
  cv2.line(img, pt3, pt0, cor, tam)
  return img
```
